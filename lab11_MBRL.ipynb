# Gaussian Process

```python
import torch
import gpytorch
import numpy as np

# ============================================================
# ---- Base Exact GP model -------------------
# ============================================================

class ExactGPModel(gpytorch.models.ExactGP):
    def __init__(self, train_x, train_y, likelihood, kernel='RBF', ard_dims=None):
        super().__init__(train_x, train_y, likelihood)
        self.mean_module = gpytorch.means.ConstantMean()

        # Choose kernel type
        if kernel == 'RBF':
            base_kernel = gpytorch.kernels.RBFKernel(ard_num_dims=ard_dims)
        elif kernel == 'Matern':
            base_kernel = gpytorch.kernels.MaternKernel(nu=2.5, ard_num_dims=ard_dims)
        elif kernel == 'RQ':
            base_kernel = gpytorch.kernels.RationalQuadraticKernel(ard_num_dims=ard_dims)
        else:
            raise ValueError(f"Unsupported kernel type: {kernel}")

        self.covar_module = gpytorch.kernels.ScaleKernel(base_kernel)

    def forward(self, x):
        mean_x = self.mean_module(x)
        covar_x = self.covar_module(x)
        return gpytorch.distributions.MultivariateNormal(mean_x, covar_x)


# ============================================================
# ---- Individual GP Manager: handles data + normalization ----
# ============================================================

class GPManager:
    def __init__(self, kernel='RBF', lr=0.03, iters=1000):
        self.kernel = kernel
        self.lr = lr
        self.iters = iters
        self.trained = False
        self.X_train = []
        self.Y_train = []

        self.likelihood = None
        self.model = None


    # ----------------------------- #
    #        FIT / INITIAL TRAIN    #
    # ----------------------------- #
    def fit(self, X, Y):
        X = torch.tensor(X, dtype=torch.float32)
        Y = torch.tensor(Y, dtype=torch.float32)

        self.X_train = X.clone()
        self.Y_train = Y.clone()

        self.retrain()

    def retrain(self):
        # Normalize the input/output and then train the model
        self._compute_normalization()
        self._train_model()

    # ----------------------------- #
    #       ADD NEW DATA POINTS     #
    # ----------------------------- #
    def add_data(self, X_new, Y_new, retrain=True):
        """
        Add new data (single or batch) to GP training set.
        Optionally retrain the model.
        """
        X_new = torch.tensor(X_new, dtype=torch.float32)
        Y_new = torch.tensor(Y_new, dtype=torch.float32).flatten()  # ensure 1D

        # Also flatten stored Y to ensure shape consistency
        if self.Y_train.ndim > 1:
            self.Y_train = self.Y_train.flatten()

        # Append new data
        self.X_train = torch.cat([self.X_train, X_new], dim=0)
        self.Y_train = torch.cat([self.Y_train, Y_new], dim=0)

        if retrain:
            self.retrain()

    # ----------------------------- #
    #         INTERNAL UTILS        #
    # ----------------------------- #
    def _compute_normalization(self):
        self.X_mean, self.X_std = self.X_train.mean(0), self.X_train.std(0)
        self.Y_mean, self.Y_std = float(self.Y_train.mean(0)), float(self.Y_train.std(0))

        Xn = (self.X_train - self.X_mean) / self.X_std
        Yn = (self.Y_train - self.Y_mean) / self.Y_std

        self.Xn, self.Yn = Xn, Yn

    def dataset(self):
        # Extract training data (already stored inside the wrapper)
        X_train = self.X_train.numpy()
        Y_train = self.Y_train.numpy()
        return X_train, Y_train


    def _train_model(self):
        self.likelihood = gpytorch.likelihoods.GaussianLikelihood()
        self.likelihood.noise_covar.initialize(noise=1e-3)
        self.model = ExactGPModel(self.Xn, self.Yn, self.likelihood,
                                  kernel=self.kernel, ard_dims=self.X_train.shape[-1])

        # -------------------------------------------------
        # Train the model
        # -------------------------------------------------
        self.train_gp(self.model, self.likelihood, self.Xn, self.Yn)
        self.trained = True

    def train_gp(self, model, likelihood, x, y):
        model.train(); likelihood.train()
        opt = torch.optim.Adam(model.parameters(), lr=self.lr)
        mll = gpytorch.mlls.ExactMarginalLogLikelihood(likelihood, model)
        for _ in range(self.iters):
            opt.zero_grad()
            out = model(x)
            loss = -mll(out, y)
            loss.backward()
            opt.step()
        model.eval(); likelihood.eval()

    def predict(self, X):
        if not self.trained:
            raise RuntimeError("GP has not been trained yet.")

        # Normalize the input
        X = torch.as_tensor(X, dtype=torch.float32)
        Xn = (X - self.X_mean) / self.X_std
        with torch.no_grad(), gpytorch.settings.fast_pred_var():
            pred = self.likelihood(self.model(Xn))

            # Unonormalize the output
            mean = pred.mean.cpu().numpy() * self.Y_std + self.Y_mean
            var = pred.variance.cpu().numpy() * (self.Y_std ** 2)
        # Return the unnormalized output
        return mean, var
```

# Setup environment and collect initial data

```python
import gymnasium as gym
# import gym
import numpy as np
import torch
import matplotlib.pyplot as plt

# --------------------------------------------------------------
# 1.  Environment setup
# --------------------------------------------------------------
env = gym.make("MountainCarContinuous-v0", max_episode_steps=500)
n_state = env.observation_space.shape[0]    # [position, velocity]
n_action = env.action_space.shape[0]        # 1 continuous action
n_output = 2                                # Δpos, Δvel


# --------------------------------------------------------------
# 2.  Collect initial random data
# --------------------------------------------------------------
data = {'X': [], 'Y': []}

obs, _ = env.reset(seed=0)
for _ in range(200):  # collect 500 transitions
    a = env.action_space.sample()  # random continuous action in [-1,1]
    x_next, r, term, trunc, _ = env.step(a)
    data['X'].append(np.concatenate([obs, a]))   # [pos, vel, action]
    data['Y'].append(x_next - obs)               # Δstate
    obs = x_next
    if term or trunc:
        obs, _ = env.reset()

X = np.array(data['X'])[::3]
Y = np.array(data['Y'])[::3]
print(f"Collected {len(X)} transitions.")


# --------------------------------------------------------------
# 3.  Train GP dynamics models
# --------------------------------------------------------------
gps = [GPManager(kernel='RBF', iters=300) for _ in range(n_output)]

for d in range(n_output):
    gps[d].fit(X, Y[:, d])
    print(f"Trained GP for Δstate[{d}] with {len(X)} samples.")
```

# MPC


```python
import numpy as np

# --------------------------------------------------------------
# MPC Parameters
# --------------------------------------------------------------
H = 15  # Prediction Horizon
K = 100 # Number of action sequences to sample

a_min = env.action_space.low[0]   # -1.0
a_max = env.action_space.high[0]  # 1.0

GOAL_POSITION = 0.5

# --------------------------------------------------------------
# Cost Function
# --------------------------------------------------------------
def cost_function(states):
    """Calculates the cost for a given trajectory of states."""
    # Penalize distance from goal (position 0.5)
    position_cost = 1000 * (GOAL_POSITION - states[:, 0])**2

    # Penalize high velocity (especially negative velocity)
    # We want to reach the goal, so positive velocity towards goal is good
    velocity_cost = 1 * states[:, 1]**2

    # Further penalize if car is behind the goal position
    # if car_state[0] < GOAL_POSITION:
    #     position_cost += 10000 * (GOAL_POSITION - car_state[0])**2

    # Sum of costs over the trajectory, weighted by time or discount factor can be added
    total_cost = np.sum(position_cost + velocity_cost)
    return total_cost


# --------------------------------------------------------------
# MPC Action Function
# --------------------------------------------------------------
def mpc_action(obs, gps, H, K, a_min, a_max):
    """Selects an action using Model Predictive Control with GP dynamics."""
    current_state = obs
    best_action_sequence = None
    min_total_cost = float('inf')

    # Store all sampled action sequences and their costs
    all_action_sequences = np.random.uniform(a_min, a_max, size=(K, H, n_action))

    for k in range(K):
        sampled_actions = all_action_sequences[k] # shape (H, n_action)

        # Simulate trajectory for this action sequence
        simulated_trajectory = [current_state]
        next_state = current_state

        for h in range(H):
            action = sampled_actions[h]

            # Predict state change using GPs
            gp_input = np.concatenate([next_state, action]).reshape(1, -1)

            delta_state_mean = np.array([gps[i].predict(gp_input)[0] for i in range(n_output)]).flatten()

            # Update state for simulation (using mean prediction for simplicity, ignoring variance for Ex1)
            next_state = next_state + delta_state_mean

            # Ensure states stay within environment bounds (optional, but good practice)
            next_state[0] = np.clip(next_state[0], env.observation_space.low[0], env.observation_space.high[0])
            next_state[1] = np.clip(next_state[1], env.observation_space.low[1], env.observation_space.high[1])

            simulated_trajectory.append(next_state)

        simulated_trajectory = np.array(simulated_trajectory[1:]) # Exclude initial state

        # Calculate cost for the simulated trajectory
        current_total_cost = cost_function(simulated_trajectory)

        # Check if this is the best sequence found so far
        if current_total_cost < min_total_cost:
            min_total_cost = current_total_cost
            best_action_sequence = sampled_actions

    # Return the first action of the best sequence
    if best_action_sequence is not None:
        return best_action_sequence[0]
    else:
        # Fallback if no sequence was found (shouldn't happen with K > 0)
        return np.random.uniform(a_min, a_max, size=(n_action,))

print("MPC parameters and mpc_action function defined.")
```

```python
import matplotlib.pyplot as plt

# --------------------------------------------------------------
# Simulation Loop
# --------------------------------------------------------------

env = gym.make("MountainCarContinuous-v0") # Re-initialize env to ensure fresh start
obs, _ = env.reset(seed=42) # Start with a fixed seed for reproducibility

positions = [obs[0]]
velocities = [obs[1]]
rewards = []

terminated = False
truncated = False
steps = 0
MAX_STEPS = 500 # Maximum steps per episode

print("Starting MPC simulation...")

while not terminated and not truncated and steps < MAX_STEPS:
    action = mpc_action(obs, gps, H, K, a_min, a_max)
    next_obs, reward, terminated, truncated, _ = env.step(action)

    obs = next_obs
    positions.append(obs[0])
    velocities.append(obs[1])
    rewards.append(reward)
    steps += 1

    if steps % 50 == 0:
        print(f"Step: {steps}, Position: {obs[0]:.4f}, Velocity: {obs[1]:.4f}, Action: {action[0]:.4f}")

env.close()

print(f"Simulation finished in {steps} steps. Reached goal: {obs[0] >= GOAL_POSITION}")

# --------------------------------------------------------------
# Plotting Position vs. Time
# --------------------------------------------------------------

plt.figure(figsize=(10, 6))
plt.plot(range(len(positions)), positions, label='Car Position')
plt.axhline(y=GOAL_POSITION, color='r', linestyle='--', label='Goal Position')
plt.xlabel('Time Steps')
plt.ylabel('Position')
plt.title('Mountain Car Position over Time using MPC with GP Dynamics')
plt.grid(True)
plt.legend()
plt.show()
```

Including uncertainty in a model’s predictions forces the model to express how sure it is about each outcome rather than providing a single fixed estimate. This extra information helps prevent overconfidence

Accounting for uncertainty often improves performance because it lets algorithms weigh information more intelligently. This leads to more stable training, and ultimately more accurate and trustworthy results, especially in real-world settings where the data are messy, incomplete, or shifting over time.
